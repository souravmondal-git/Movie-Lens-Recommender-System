```python
from surprise import SVD, NMF, SVDpp, CoClustering, KNNBaseline, KNNWithZScore
```

```python
from surprise import Dataset, Reader
```

```python
reader = Reader(rating_scale=(1, 5))
```

```python
data = Dataset.load_from_df(ratings1, reader=reader)
```

```python
from surprise.model_selection import train_test_split
```

```python
trainset, testset = train_test_split(data, test_size=.25)
```

```python
from surprise.model_selection import cross_validate
```

```python
from surprise import accuracy
```

```python
import re
```

```python
algos = [KNNBaseline(), KNNWithZScore(), NMF(), SVD(), SVDpp(), CoClustering()]
metr = []
for alg in algos:
    tmp = cross_validate(alg, data=data, measures=['RMSE'], cv=3)
    tmp_alg = re.sub("\(\)","",str(alg))
    
    tmp_alg = tmp_alg.split(".")
    alg_name = tmp_alg[-1]
    alg_name = alg_name.split(" ")[0]
    metr.append((alg_name,tmp['test_rmse'].mean()))
    
table_rmse=pd.DataFrame(metr, columns=['Algos', 'RMSE'])
table_rmse
```

```python
svdpp = SVDpp()
```

```python
svdpp.fit(trainset)
```

```python
rating_matrix
```

```python
svdpp.predict(uid=320, iid=6).est
```

```python
predictions = svdpp.test(testset)
accuracy.rmse(predictions)
```

```python
def top_five_reco(user):
    list_of_ratngs=[]
    top5movs = []
    for mv in all_movies:
        if mv not in list(ratingdf.loc[user].index):
            rts = (mv, svdpp.predict(uid=user, iid=mv).est)
            list_of_ratngs.append(rts)
    list_of_ratngs.sort(key= lambda x : x[1], reverse=True)
    for i in range(5):
        top5movs.append(list_of_ratngs[i][0])
    return mov.loc[top5movs]
        
    
    
```
